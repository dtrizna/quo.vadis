```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
#import seaborn as sns

fig, ax = plt.subplots(1,2, figsize=(12,8))

img_unp = mpimg.imread("brbot_unpacked_same_size.png")
img_pck = mpimg.imread("brbot_packed_IAT_pestudio.png")
ax[0].imshow(img_pck)
ax[0].set_title("Packed sample IAT")
ax[1].imshow(img_unp)
ax[1].set_title("Un-packed sample IAT")
[(x.set_xticks([]),x.set_yticks([])) for x in ax]
plt.subplots_adjust(wspace=0, hspace=0)
```
