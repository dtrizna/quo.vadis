```python
import speakeasy
import os
import pandas as pd
```

```python
FILENAME_PATTERN = "b43"
MALWARE_PATH = "testfiles/"
file = [x for x in os.listdir(MALWARE_PATH) if FILENAME_PATTERN in x][0]
print(file)
```

```python
se = speakeasy.Speakeasy()
module = se.load_module(MALWARE_PATH+file)
se.run_module(module)
```

```python
print([x for x in dir(se) if "__" not in x])
```

```python
report = se.get_report()
report.keys()
```

```python
aa = pd.json_normalize(report)
aa
```
