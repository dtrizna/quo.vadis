```python
import os
import time
import subprocess
```

```python
defaultdir = "spc-karafiat-pen-run-1"
if not os.path.exists(defaultdir):
    subprocess.check_output("7za x -ptest smbmap_enum_enc.7z".split())
```

```python
handle = open("dataset_smbmap.txt", "a")

files = os.listdir(defaultdir)
l = len(files)
for i,file in enumerate(files):
    print(f"{i}\{l}", end=" ")
    with open(defaultdir+"/"+file) as f:
        data = f.readlines()
    
    for line in data:
        if line:
            path = " ".join(line.split()[:-1]).replace(",", "")
            if "." in path[-7:] and \
                "Error during listing" not in path \
                and "BillingBackend" not in path \
                and "CxNotifications" not in path:
                if path:
                    handle.write(path+"\n")
    
handle.close()
```
