```python
import speakeasy
import pandas as pd
import numpy as np
import os
import time
import pickle

MALWARE_PATH = "../../../data/pe.dataset/"
X86_PATH = MALWARE_PATH + "PeX86Exe/"
X86_RANSOMWARE = X86_PATH + "ransomware/"
X86_CLEAN = X86_PATH + "clean/"
X86_TROJAN = X86_PATH + "trojan/"
X86_DROPPER = X86_PATH + "dropper/"
```

```python
with open("reports_PeX86Exe_dropper_40_500.pickle", "rb") as fhandle:
    reports = pickle.load(fhandle)
```

```python
def parse_report(report):
    aa = pd.json_normalize(report)
    ab = pd.json_normalize(aa["entry_points"].iloc[0])
    try:
        if ab["apis"].any():
            #print(ab["apis"])
            ac = pd.json_normalize(ab["apis"].iloc[0])
        else:
            ac = []
    except AttributeError:
        ac = []
    return aa, ab, ac

bs = []
success = 0
unsupported_apis = []
for i,f in enumerate(reports.keys()):
    a, b, c = parse_report(reports[f])
    if "error.type" in b.keys():
        apis = np.sum([len(x["apis"]) for _,x in b.iterrows()])
        if apis > 5:
            success += 1
        if b["error.type"].iloc[0] == "unsupported_api":
            unsupported_apis.append(b["error.api_name"].iloc[0])
            print(f," : unsupported_api :" , b["error.api_name"].iloc[0], f": len of api seq. before fail: {len(b['apis'].iloc[0])}")
        else:
            print(f, " : ", b["error.type"].iloc[0], f" len of api seq. before fail: {apis}")
    else:
        success += 1

print(f"success ratio: {success*100/len(reports)} %")
```
